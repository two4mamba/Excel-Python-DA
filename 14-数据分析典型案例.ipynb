## 14.1 利用Python实现报表自动化

```text
        |  上月同期     去年同期    环比    同比
----------------------------------------------------
销售额  |
客流量  |
客单价  |
```
指标说明<br>
- 销售额 = 单价 * 销量
- 客流量 = 订单ID去重计数
- 客单价 = 销售额/客流量
- 本月 = 2018年2月
- 上月 = 2018年1月
- 去年同期 = 2017年2月


### 导入源数据

````python
import pandas as pd
data = pd.read_csv('data/order-14.1.csv',encoding='gbk',parse_dates=['成交时间'])
data.head()
````

上面`parse_dates`参数表示将数据解析为时间格式

### 计算本月相关指标

````python
from datetime import datetime
This_month = data[(data['成交时间'] >= datetime(2018, 2, 1))
                  & (data['成交时间'] <= datetime(2018, 2, 28))]

# 计算本月销售额
sales1 = (This_month['单价']*This_month['销量']).sum()

#计算本月客流量
traffic1 = This_month['订单ID'].drop_duplicates().count()

# 计算本月客单价
s_t1 = sales1/traffic1

print(f"本月销售额为:{sales1},客流量为：{traffic1},客单价为：{s_t1}")
````

### 计算上月相关指标

````python
last_month = data[(data['成交时间'] >= datetime(2018, 1, 1))
                  & (data['成交时间'] <= datetime(2018, 1, 31))]
# 计算上月销售额
sales2 = (last_month['单价']*last_month['销量']).sum()

#计算本月客流量
traffic2 = last_month['订单ID'].drop_duplicates().count()

# 计算本月客单价
s_t2 = sales2/traffic2

print(f"本月销售额为:{sales2},客流量为：{traffic2},客单价为：{s_t2}")
````

### 计算去年同期相关指表

````python
same_month = data[(data['成交时间'] >= datetime(2017, 2, 1))
                  & (data['成交时间'] <= datetime(2017, 2, 28))]

# 计算本月销售额
sales3 = (same_month['单价']*same_month['销量']).sum()

#计算本月客流量
traffic3 = same_month['订单ID'].drop_duplicates().count()

# 计算本月客单价
s_t3 = sales3/traffic3

print(f"本月销售额为:{sales3},客流量为：{traffic3},客单价为：{s_t3}")
````

### 利用函数提高编码效率

如上，其中销售鹅、客流量、客单价只是每个月的月份不同，计算方法是一样的，因此可以可以用函数替代，把每个月的数据作为参数传递给函数，然后返回值为销售额、客流量、客单价

````python
def get_month_keys(data):
    sales = (data['单价'] * data['销量']).sum()
    traffic = data['订单ID'].drop_duplicates().count()
    s_t = sales / traffic
    return sales, traffic, s_t
````

````python
sales1, traffic1, s_t1 = get_month_keys(This_month)
sales2, traffic2, s_t2 = get_month_keys(last_month)
sales3, traffic3, s_t3 = get_month_keys(same_month)
````

### 将三个时间段的指标合并

````python
report = pd.DataFrame([[sales1, sales2, sales3],
                       [traffic1, traffic2, traffic3], [s_t1, s_t2, s_t3]],
                      columns=['本月累计', '上月同期', '去年同期'],
                      index=['销售量', '客流量', '客单价'])

report
````

### 添加同比和环比字段

````python
report['环比'] = report['本月累计'] / report['上月同期'] - 1
report['同比'] = report['本月累计'] / report['去年同期'] - 1

report
````

### 将结果文件导出到本地

````python
report.to_csv('data/order.csv',encoding='utf-8-sig')
````

## 14.3 假如你是某连锁超市的数据分析师

假如你是一家连锁超市的数据分析师，以下几个小节讲的问题可能会是你经常需要关注的。数据源如下:

````python
import pandas as pd
saleData = pd.read_csv('data/order-14.3.csv',
                       parse_dates=['成交时间'],
                       encoding='gbk')
saleData
````

### 14.3.1 哪些类别的商品比较畅销

要看哪些类别的商品比较畅销，只要将订单表中的数据按照类别ID进行分组，然后对分组后的销量求和，就会得到每一类在一段时间内的销量:

````python
saleData.groupby('类别ID')['销量'].sum().reset_index()
````

运行上面的代码得到了所有类别在一段时间内对应的销量，**我们想看销量最好的前10个类别，就要先对销量做一个降序排列，然后取前10行**即可:

````python
saleData.groupby('类别ID')['销量'].sum().reset_index().sort_values(
    by='销量', ascending=False).head(10)
````

### 14.3.2 哪些商品比较畅销

计算哪些商品比较畅销，其实与计算哪些类别比较畅销的逻辑一致，**上面用了数据分组，这次用数据透视表来计算哪些商品比较畅销**，同样取前10名的商品。

````python
pd.pivot_table(saleData, values='销量', index='商品ID',
               aggfunc='sum').reset_index().sort_values(
                   by='销量', ascending=False).head(10)
````

### 14.3.3 不同门店的销售额占比

商品畅销程度直接用销量来表示即可，销售额等于销量乘单价，订单表中没有销售额字段，**所以需要新增一个销售额字段。新增字段以后，按照门店编号进行分组，对分组后的营业额求和运算，最后计算不同门店的销售额占比**

````python
saleData['销售额'] = saleData['单价'] * saleData['销量']
saleData.groupby('门店编号')['销售额'].sum().reset_index()
````

````python
# 计算销售额占比
saleData.groupby('门店编号')['销售额'].sum() / saleData['销售额'].sum()
````

````python
%config InlineBackend.figure_format = 'svg'

# 绘制饼图
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']

(saleData.groupby('门店编号')['销售额'].sum() / saleData['销售额'].sum()).plot.pie()
````

### 14.3.4 哪些时间段是超市的客流高峰期

现在我们想知道一天中什么时间段（哪几个小时）是高峰期，要想找出高峰期时间段，**需要知道每个时间段对应的客流量，但是订单表中的成交时间既有日期又有时间，我们需要从中提取出小时数，这里依然用订单ID去重计数代表客流量**。

````python
# 利用自定义时间格式函数strftime提取小时数
saleData['小时'] = saleData['成交时间'].map(lambda x: int(x.strftime('%H')))

# 对小时和订单去重
traffic = saleData[['小时', '订单ID']].drop_duplicates()

# 求每小时的客流量
traffic.groupby('小时')['订单ID'].count()
````

````python
%config InlineBackend.figure_format = 'svg'

# 绘制每个时段的折线图
traffic.groupby('小时')['订单ID'].count().plot()
````

上述代码中之所以要对小时和订单进行去重处理，是因为我们用的订单表是以商品ID为主键的，在一个小时内可能会出现多个相同的订单
ID，这些订单ID来自同一个人，所以算作一个人。

分小时客流量折线图中可以看出8点到10点是超市一天中的销售高峰期，17点到19点又有一个销售小高峰。

## 14.4 假如你是某银行的数据分析师

**假如你是某银行的数据分析师，那么坏账率肯定是你日常需要关注的重点指标，坏账率的高低主要会受哪些因素影响呢？**现在有一份历史借款人员明细表，通过这份历史记录来看一下坏账率都会受哪些因素的影响，记录表如下：

````python
loanData = pd.read_csv('data/loan.csv', encoding='gbk')
loanData.info()
````

````python
loanData.head(10)
````

### 14.4.1 是不是收入越高的人坏账率越低

按理来说收入越高的人越不缺钱，坏账率应该越低，那么实际情况是什么样的呢？

我们通过数据来看一下。在借款人员明细表的基本信息中，**月收入是有缺失值的，所以在具体分析以前，我们要先做一个缺失值处理。这里选择用均值填充的方法**,如下：

````python
loanData = loanData.fillna({'月收入': loanData['月收入'].mean()})
loanData.info()
````

通过`loanData.info()`函数可以看到，月收入已经没有缺失值了，可以正式分析了。

**因为月收入属于连续值，对于连续值进行分析时，我们一般都会将连续值离散化，就是将连续值进行区间切分，分成若干类别**：

````python
cut_bins = [0, 5000, 10000, 15000, 20000, 100000]
income_cut = pd.cut(loanData['月收入'], cut_bins)
income_cut
````

区间切分好以后就可以看每个区间内的坏账率，坏账率又该怎么计算呢？

**坏账率就是所有借款用户中逾期不还用户的占比**。逾期不还用户的好坏客户字段标记为1，非逾期不还用户的好坏客户字段标记为0。
<br>坏账率就等于好坏客户字段之和（坏账客户数）与好坏客户字段的计数（所有借款用户）的比值。

````python
all_income_user = loanData['好坏客户'].groupby(income_cut).count()
bad_income_user = loanData['好坏客户'].groupby(income_cut).sum()
bad_rate = bad_income_user / all_income_user
bad_rate
````

````python
# 绘制月收入与坏帐率的关系图
bad_rate.plot.bar()
````

如上图所示当月收入在1万元以下时，收入越高，坏账率越低，当月收入超过1.5万元时，坏账率又出现了上涨。所以并不完全是月收入越高，坏账率越低，只是在一定范围内，月收入越高坏账率会越低。

### 14.4.2 年龄和坏账率有什么关系

年龄和坏账率有什么关系呢？是不是年龄越大消费越理性，对信用越看重，坏账率越低呢？

年龄也是连续值，也是用连续值离散化方式处理，代码如下

````python
age_cut = pd.cut(loanData['年龄'],6)
all_age_user = loanData['好坏客户'].groupby(age_cut).count()
bad_age_user = loanData['好坏客户'].groupby(age_cut).sum()
bad_age_rate = bad_age_user / all_age_user
bad_age_rate
````

````python
# 绘制年龄与坏帐关系图
bad_age_rate.plot.bar()
````

18岁以下的人的坏账率为0,18~36岁的人的坏账率最高，超过36岁的人随着年龄的增加，坏账率呈下降趋势。

### 14.4.3 家庭人口数量和坏账率有什么关系

家庭人口数量和坏账率有什么关系呢？是家庭人口越多，负担越重，坏账率越高？还是家庭人口越多，劳动力越多，坏账率越低呢？我
们具体看一下家庭人口数量和坏账率的关系。

虽然人口数量也是连续值，但是因为数值不是很大，所以我们就当作离散值处理，不进行区间切分。

````python
all_user = loanData.groupby('家属数量')['好坏客户'].count()
bad_user = loanData.groupby('家属数量')['好坏客户'].sum()
bad_rate = bad_user / all_user
bad_rate
````

````python
# 绘制家属数量和坏帐关系图
bad_rate.plot()
````
